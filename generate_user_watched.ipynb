```python
import pandas as pd

file_path = './data/cleaned/anime_data_cleaned.csv'
anime_data = pd.read_csv(file_path)

anime_data.head()
```

```python
import random

num_users = 10000
user_ids = [f"{i+1}" for i in range(num_users)]

titles = anime_data['title'].tolist()

watched_data = {
    "user_id": user_ids,
    "titles_watched": [
        ", ".join(random.sample(titles, random.randint(1, 20))) for _ in range(num_users)
    ]
}

watched_df = pd.DataFrame(watched_data)

output_path = './data/cleaned/anime_watched_data.csv'
watched_df.to_csv(output_path, index=False)

output_path
```
