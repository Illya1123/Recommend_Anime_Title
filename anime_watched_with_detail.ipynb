```python
import pandas as pd
import os
```

```python
anime_watched = pd.read_csv('./data/cleaned/anime_watched_data.csv')
anime_data = pd.read_csv('./data/cleaned/anime_data_cleaned.csv')

anime_watched.head()
```

```python
anime_data.head()
```

```python
# Tách các title trong cột titles_watched của anime_watched_data
anime_watched['titles_watched'] = anime_watched['titles_watched'].apply(lambda x: x.split(', '))

# Mở rộng data anime_watched_data để mỗi hàng là một user và một title mà họ đã xem
anime_watched_expanded = anime_watched.explode('titles_watched').reset_index(drop=True)
```

```python
# Kết hợp dữ liệu anime_watched_expanded với anime_data_cleaned dựa trên title
merged_data = pd.merge(anime_watched_expanded, anime_data, left_on='titles_watched', right_on='title', how='left')
```

```python
cleaned_data = merged_data.dropna()
```

```python
print(merged_data[['user_id', 'titles_watched', 'link', 'image', 'episode', 'rating', 'views', 'quality', 'year', 'genres', 'description', 'season', 'type', 'rate', 'nums_of_vote']])
```

```python
output_dir = 'data/proceeded'
```

```python
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
output_path = os.path.join(output_dir, 'anime_watched_with_details.csv')
merged_data[['user_id', 'titles_watched', 'link', 'image', 'episode', 'rating', 'views', 'quality', 'year', 'genres', 'description', 'season', 'type', 'rate', 'nums_of_vote']].to_csv(output_path, index=False)

print(f"File CSV đã được xuất ra tại {output_path}")
```
